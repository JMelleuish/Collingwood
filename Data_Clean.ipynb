```python
import pandas as pd
```

## Renaming Columns

```python
df = df.rename(columns={"Home/Away": "HomeAway", "Final Score": "FinalScore", "Actual Crowd": "ActualCrowd", "Win/Loss":"WinLoss", "Ladder Position":"LadderPosition", "Games Won in last 5 played": "Last5"})
df
```

## Sorting inappropriate values
### Team

```python
df.Team.value_counts()
```

### Round

```python
df.Round.value_counts()
```

All values appear to be reasonable except round 0, we now look into whats happeing here:

```python
df[df['Round'] == 0]
```

All of the above instances have no meanful data/ do not represent an actual match that has taken place, thus it is reasonable to remove these instances

```python
df = df[df['Round'] != 0]
df.Round.value_counts()
```

### Home/Away

```python
df.HomeAway.value_counts()
```

Given that a team neither wins or losses if they have the bye, we can remove all instances of teams with the bye

```python
df = df[df['HomeAway'] != 'Bye']
df.HomeAway.value_counts()
```

### Final Score

```python
df[df['FinalScore'] < 0] # Not possible
```

```python
df[df['FinalScore'] > 239] # Highest VFL/AFL score in history
```

```python
df['FinalScore'] = df['FinalScore'].fillna('NaN')
df[df['FinalScore'] == 'NaN']
```

All of the above instances seem to be fake records, thus we will remove them

```python
df = df[df['FinalScore'] != 'NaN']
```

### Date

```python
df[df['Date'] == min(df['Date'])]
```

```python
df["Date"] = df["Date"] - 40626
```

Day 0 occurs on the 24th of march 2011, thus we can add the days after

```python
basedate = pd.Timestamp('2011-3-24')
df['Date'] = pd.to_timedelta(df['Date'], unit='d') + basedate
```

### Day

```python
df.Day.value_counts()
```

### Time

```python
df['Time'] = round(df['Time']*24) # Rounding here because the minutes past the hour is assumed to be independent of a win or loss
```

```python
df.Time.value_counts()
```

Numbers greater than 12 are assumed to be in 24 hour time and the rest not as no afl game has ever been played before midday

```python
df.loc[df["Time"]>=12, "Time"] -= 12
df.Time.value_counts()
```

```python
df[df['Time'] == 9]
```

9 is plausible for the time difference between Melbourne and Perth

```python
df[df['Time'] == 0]
```

All of these places are outside of victoria, hence due to time differences these times are plausible

### Venue

```python
df.Venue.value_counts()
```

Some of the above values are duplicates that have been spelt wrong, hence we adjust the following:

```python
df['Venue'] = df['Venue'].str.strip() # stripping white spaces infront of names
df["Venue"] = df["Venue"].replace("Kardina Park", "Kardinia Park")
df["Venue"] = df["Venue"].replace("Kardinia", "Kardinia Park")
df["Venue"] = df["Venue"].replace("GMHBA Stadium", "Kardinia Park")
df["Venue"] = df["Venue"].replace("Marvel Stadium", "Docklands")
df["Venue"] = df["Venue"].replace("S.C.G.", "SCG")
df["Venue"] = df["Venue"].replace("Optus Stadium", "Perth Stadium")
df["Venue"] = df["Venue"].replace("Football", "Football Park") # Assumption given it is the only oval with football in its name
df["Venue"] = df["Venue"].replace("Sydeny Show GroundS", "Sydney Showground")
df["Venue"] = df["Venue"].replace("GIANTS Stadium", "Sydney Showground")
df["Venue"] = df["Venue"].replace("Sydney Show Grounds", "Sydney Showground")
df["Venue"] = df["Venue"].replace("York", "York Park")
df["Venue"] = df["Venue"].replace("TIO Stadium", "Marrara Oval")
df["Venue"] = df["Venue"].replace("Marrara", "Marrara Oval")
df["Venue"] = df["Venue"].replace("UTAS Stadium", "York Park")
df["Venue"] = df["Venue"].replace("Manuka", "Manuka Oval")
df["Venue"] = df["Venue"].replace("Varrara", "Carrara")
df["Venue"] = df["Venue"].replace("Metricon Stadium", "Carrara")
df["Venue"] = df["Venue"].replace("Adelaide", "Adelaide Oval")
df["Venue"] = df["Venue"].replace("TIO Traeger Park", "Traeger Park")
df["Venue"] = df["Venue"].replace("Jianwan Stadium", "Jiangwan Stadium")
df["Venue"] = df["Venue"].replace("Cazaly's", "Cazaly's Stadium")
df["Venue"] = df["Venue"].replace("Bellervie Oval", "Bellerive Oval")
df["Venue"] = df["Venue"].replace("Mars Stadium", "Eureka Stadium")
df["Venue"] = df["Venue"].replace("University of Tasmania Stadium", "York Park")
df["Venue"] = df["Venue"].replace("Blundstone Arena", "Bellerive Oval")
df.Venue.value_counts()
```

```python
df[df['Venue'] == "Stadium"]
```

Given all of these are with sydney as the home team we can assume that they are from stadium australia:

```python
df["Venue"] = df["Venue"].replace("Stadium", "Stadium Australia")
df[df['Venue'] == "Park"]
```

Given this was a Geelong home game it is safe to assume it was at kadinia park:

```python
df["Venue"] = df["Venue"].replace("Park", "Kardinia Park")
df.Venue.value_counts()
```
